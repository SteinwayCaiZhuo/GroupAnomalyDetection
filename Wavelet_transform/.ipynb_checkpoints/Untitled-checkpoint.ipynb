```python
import pywt
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def plot_np(a, scale=False):
    max_a = np.max(a)
    min_a = np.min(a)
    if(max_a == min_a):
        print("img == ", min_a)
        scale = False
        
    b=a
    if(scale):
        b = (a-min_a)/(max_a-min_a)*255.
    
    plt.figure()
    plt.imshow(b)
    plt.show()
```

```python
#a = np.random.random(30*30).reshape(30,30)*0+100
h = 300
w = 300
a = (np.arange(h*w).reshape(h,w)/h/w*255.0).astype(int)

for i in range(a.shape[0]):
    for j in range(a.shape[1]):
        a[i][j] = (i-a.shape[0]/2)*(i-a.shape[0]/2) - (j-a.shape[1]/2)*(j-a.shape[1]/2)

plot_np(a, scale=True)
```

```python
ca = pywt.wavedec2(a, "Haar", level=8)
```

```python
for i in ca:
    for j in i:
        if(len(j.shape)==2):
            plot_np(j)
```
