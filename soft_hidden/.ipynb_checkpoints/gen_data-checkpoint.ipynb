```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas
```

```python
N = 1000
p = 1
```

```python
x = np.zeros((N, p))
```

```python
mul = 0.01
for i in range(N-1):
    if np.random.random()>0.9:
        mul*=(-1)
    if x[i] > 1:
        mul= -np.abs(mul)
    elif x[i] < -1:
        mul = np.abs(mul)
    x[i+1] = np.cos(x[i])*(mul)+x[i]
```

```python
plt.plot(x)
```

```python
# The ability of a dnn to learn cos function
```

```python
import tensorflow as tf
```

```python
x = np.linspace(-10,10,1000)
```

```python
y = np.cos(x)
```

```python
model = tf.keras.models.Sequential([
    #tf.keras.layers.Flatten(input_shape=(1,dimension)),
    tf.keras.layers.Dense(1, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.relu),
    #tf.keras.layers.Dense(100, activation=tf.nn.tanh),
    tf.keras.layers.Dense(100, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.relu),
    
#     tf.keras.layers.Dense(1, activation=tf.nn.relu),
#     #tf.keras.layers.Dense(50, activation=tf.nn.relu),
    
#     tf.keras.layers.Dense(1, activation=tf.nn.relu),
#     tf.keras.layers.Dense(1, activation=tf.nn.relu),
    tf.keras.layers.Dense(1)
])
#adam = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)
model.compile(optimizer='Adam',
              loss=tf.keras.losses.mean_squared_error,
               metrics=["mse"])
model.fit(x.reshape(-1,1), y.reshape(-1,1), epochs=50)
```
