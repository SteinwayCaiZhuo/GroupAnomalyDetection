```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
swat = pd.read_csv("../../data/SWaT/SWaT_Dataset_Attack_Separated_Process.csv")
```

```python
curr_columns = swat.columns
new_columns = swat.iloc[0,:]
```

```python
print(curr_columns)
print("len:",len(curr_columns))
```

```python
print(new_columns)
print("len:", len(new_columns))
```

```python
column_map = {}
for i in range(len(curr_columns)):
    column_map[curr_columns[i]] = new_columns[i]
```

```python
swat2 = swat.rename(columns=column_map)
```

```python
swat3 = swat2.iloc[1:,:]
```

```python
n_a = swat3["Normal/Attack"].to_numpy().astype("double")
```

```python
swat_np = swat3.iloc[:,1:-2].to_numpy().astype("double")
```

```python
swat_np.shape
```

```python
np.save("../../data/SWaT/swat_anomaly.npy", swat_np)
```

```python
np.sum(np.isnan(swat_np))
```

```python
start = []
end = []
for i in range(swat_np.shape[0]-1):
    if(swat_np[i][-1]==0 and swat_np[i+1][-1]==1):
        #print("0->1", i, i+1)
        start.append(i+1)
    elif(swat_np[i][-1]==1 and swat_np[i+1][-1]==0):
        #print("1->0", i, i+1)
        end.append(i+1)

attack_ = np.concatenate((start, end)).reshape(2, len(start)).transpose(1,0)
```

```python
print("attack: ", attack_)
```

```python
plt.figure(figsize = (20,10*(swat_np.shape[1]-1)))
range2 = range(0, swat_np.shape[0])
for i in range(swat_np.shape[1]-1):
    plt.subplot(swat_np.shape[1]-1, 1, i+1)
    plt.plot(swat_np[range2,:][:,i])
    plt.title("sensor"+str(i))
    for j in range(len(start)):
        if(start[j] in range2 or end[j] in range2):
            plt.axvspan(start[j], end[j], alpha=0.3, color='red')
```
