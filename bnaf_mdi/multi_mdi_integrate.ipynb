```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
```

```python
sorted(os.listdir())
```

```python
p1 = np.load('_p1.npy')
p2 = np.load('_p2.npy')
p3 = np.load('_p3.npy')
px = np.load('_px.npy')
```

```python
plt.figure(figsize=(10,4*4))
plt.subplot(4,1,1)
plt.plot(p1)
plt.subplot(4,1,2)
plt.plot(p2)
plt.subplot(4,1,3)
plt.plot(p3)
plt.subplot(4,1,4)
plt.plot(px)
```

```python
p1.max(), p2.max(), p3.max(), px.max()
```

```python
p1.mean(), p2.mean(), p3.mean(), px.mean()
```

```python
p1 = p1/np.std(p1)
p2 = p2/np.std(p2)
p3 = p3/np.std(p3)
px = px/np.std(px)
```

```python
plt.figure(figsize=(10,4*4))
plt.subplot(4,1,1)
plt.plot(p1)
plt.subplot(4,1,2)
plt.plot(p2)
plt.subplot(4,1,3)
plt.plot(p3)
plt.subplot(4,1,4)
plt.plot(px)
```

```python
p_max = np.max((p1,p2,p3,px), axis=0)
p_max.shape
```

```python
np.save('_p_max.npy', p_max)
```

```python
p_sum = p1+p2+p3+px
np.save('_p_sum.npy', p_sum)
```
