```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv("../data/WADI/WADI_attackdata_labelled.csv") 
```

```python
cols = list(data)
list(data)
```

```python
# Find the start and end of each ground truth attack
attack = data['attack']
heads = []
tails = []
for i in range(len(attack)):
    if attack[i] == 1:
        if attack[i-1] == 0:
            heads.append(i)
        if attack[i+1] == 0:
            tails.append(i)
print(heads)
print(tails)
```

```python
plt.figure(figsize=(12, 2))
lt2 = data['\\\\WIN-25J4RO10SBF\\LOG_DATA\\SUTD_WADI\\LOG_DATA\\2_LT_002_PV']
plt.plot(lt2, '-b')
for i in range(len(heads)):
    plt.axvspan(heads[i], tails[i], alpha=0.3, color='red')
```

```python
# Plot all time series sensors
for i in range(5, len(cols)):
    plt.figure(figsize=(15, 1.5))
    plt.title(cols[i])
    cur = data[cols[i]]
    plt.plot(cur, '-b')
    for i in range(len(heads)):
        plt.axvspan(heads[i], tails[i], alpha=0.3, color='red')
```
